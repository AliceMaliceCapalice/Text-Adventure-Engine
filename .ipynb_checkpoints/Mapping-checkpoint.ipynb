```python
from PIL import Image, ImageDraw, ImageOps
import math
#make the label optional, instead have it use the direction you input.
class Mapping:
    def __init__(self, geometry):
        self.coords = [geometry//2,geometry//2]
        self.rooms = {}
    def addRoom(self, name, direction, label="none", blue=False):
        if direction=="alone":
            #checks if the room should be unconnected or not
            if blue:
                Mapping.drawRoom(self.coords, name, (0,0,255))
            else:
                Mapping.drawRoom(self.coords, name, (0,0,0))
            self.rooms[name.strip().lower()] = self.coords
            return
        self.move("none", direction)
        self.rooms[name.strip().lower()] = self.coords
        if blue:
            Mapping.drawRoom(self.coords, name, (0,0,255))
        else:
            Mapping.drawRoom(self.coords, name, (0,0,0))
        Mapping.drawLine(self.coords, direction)
        if not label=="none":
            Mapping.addLabel(self.coords, direction, label)
    def move(self, destination, direction):
        if not destination == "none":
            #checks if there is a destination
            if destination[0] in "1234567890":
                #checks if you are using the numerical location
                for num, name in enumerate(self.rooms.keys()):
                    if num == int(destination):
                        self.coords=self.rooms[name]
                        return
            if destination in self.rooms.keys():
                #checks if you input a name for a location
                self.coords=self.rooms[destination]
                return
            else:
                print("\nYou input a non-existant room")
                return
        if(direction == "north"):
            self.coords = [self.coords[0], self.coords[1]-150]
        if(direction == "south"):
            self.coords = [self.coords[0], self.coords[1]+150]
        if(direction == "east"):
            self.coords = [self.coords[0]+150, self.coords[1]]
        if(direction == "west"):
            self.coords = [self.coords[0]-150, self.coords[1]]
        if(direction == "north-west"):
            self.coords = [self.coords[0]-150, self.coords[1]-150]
        if(direction == "north-east"):
            self.coords = [self.coords[0]+150, self.coords[1]-150]
        if(direction == "south-west"):
            self.coords = [self.coords[0]-150, self.coords[1]+150]
        if(direction == "south-east"):
            self.coords = [self.coords[0]+150, self.coords[1]+150]
    @staticmethod
    def addLabel(coords, direction, text):
        #This labels the line
        if(direction == "north"):
            draw.text((coords[0], coords[1]+70), text, fill="black")
        if(direction == "south"):
            draw.text((coords[0], coords[1]-80), text, fill="black")
        if(direction == "east"):
            draw.text((coords[0]-80, coords[1]), text, fill="black")
        if(direction == "west"):
            draw.text((coords[0]+70, coords[1]), text, fill="black")
        if(direction == "north-west"):
            draw.text((coords[0]+90, coords[1]+90), text, fill="black")
        if(direction == "north-east"):
            draw.text((coords[0]-90, coords[1]+90), text, fill="black")
        if(direction == "south-west"):
            draw.text((coords[0]+90, coords[1]-90), text, fill="black")
        if(direction == "south-east"):
            draw.text((coords[0]-90, coords[1]-90), text, fill="black")
    @staticmethod
    def drawRoom(coords, text, colour):
        draw.rectangle(xy=([(coords[0]-40, coords[1]-40),(coords[0]+40,coords[1]+40)]), fill="white", outline=(255,255,255))
        draw.rectangle(xy=([(coords[0]-40, coords[1]-40),(coords[0]+40,coords[1]+40)]), outline=colour)
        draw.text((coords[0]-30, coords[1]), text, fill="black")
    @staticmethod
    def drawLine(coords, direction):
        if direction == "north":
            draw.line([(coords[0],coords[1]+40),(coords[0],coords[1]+110)], fill="black", width=1)
        if direction == "south":
            draw.line([(coords[0],coords[1]-40),(coords[0],coords[1]-110)], fill="black", width=1)
        if direction == "east":
            draw.line([(coords[0]-40,coords[1]),(coords[0]-110,coords[1])], fill="black", width=1)
        if direction == "west":
            draw.line([(coords[0]+40,coords[1]),(coords[0]+110,coords[1])], fill="black", width=1)
        if direction == "north-west":
            draw.line([(coords[0]+40,coords[1]+40),(coords[0]+110,coords[1]+110)], fill="black", width=1)
        if direction == "north-east":
            draw.line([(coords[0]-40,coords[1]+40),(coords[0]-110,coords[1]+110)], fill="black", width=1)
        if direction == "south-west":
            draw.line([(coords[0]+40,coords[1]-40),(coords[0]+110,coords[1]-110)], fill="black", width=1)
        if direction == "south-east":
            draw.line([(coords[0]-40,coords[1]-40),(coords[0]-110,coords[1]-110)], fill="black", width=1)
        
           
#change this line to alter the image size
version = input("Do you want to use an existing map? \"yes\" or \"no\"")
if version == "yes":
    image = Image.open("File.png")
    geometry = image.size[0]
elif version == "no":
    geometry = 190*int(input("how many rooms do you plan on making?"))
    image = Image.new("RGB", (geometry,geometry), (255, 255, 255))
draw = ImageDraw.Draw(image)

card = Mapping(geometry)
name = input("What is the starting location called?")
card.addRoom(name, "alone", "none", False)
while True:
    option = input("What would you like to do?\n 1. move.\n 2. add a room adjacent to the current room.\n 3. List rooms. \n 4. Shorthand mode.\n 5. to quit and to save image.\n")
    if option in "abcdefghijklmnopqrstuvwxyz":
        print("You input an invalid option")
    else:
        match int(option):
            case 1:
                #Move to direction
                option = input("\nDo you want to 1. go somewhere, or 2. move in a direction")
                if option in "abcdefghijklmnopqrstuvwxyz":
                    print("You input an invalid option")
                else:
                    match int(option):
                        case 1:
                            #changing the current room
                            destination = input("Where do you want to go?")
                            card.move(destination, "random")
                            print(f"\nYou are now making rooms for {destination}")
                        case 2:
                            #moving in a direction
                            direction = input("What is the direction you want to move?")
                            card.move("none", direction)
                        case _:
                            print("You entered an invalid input")
            case 2:
                #add adjacent room
                name = input("What is the next room called?")
                direction = input("What is the direction of this room?")
                label = input("What do you want the label of the line to be? If you don't want it, type none")
                card.addRoom(name, direction, label, False)
            case 3:
                #list rooms
                for num, room in enumerate(card.rooms.keys()):
                    print("\n")
                    print(f"{num}. {room}")
            case 4:
                #shorthand mode, lets you type in commands instead of going through the menu.
                print("\nTo use this, you can \"add [name] [direction] [label, if using addl]\"\nyou can move somewhere with \"move to [room name]\" or \"move [direction] [optional-n times]\" or with \"move [room number]\"\nYou can list the rooms with \"list\"\nYou can add a solitary blue room with \"addb [name]\"\nAnd you can exit with \"q\" or \"quit\".\n\n")
                while True:
                    commands = input("\n>")
                    commands = commands.strip().lower()
                    split = commands.split()
                    
                    if(len(split) == 1):
                        if(split[0] == "list"):
                            #lists all the rooms
                            print("\n")
                            for num, room in enumerate(card.rooms.keys()):
                                print(f"{num}. {room}")
                        if("quit" in split or "q" in split):
                            break
                    if(len(split) >= 2):
                        if(split[0] == "move" and split[1] == "to" and len(split) >= 3):
                            #puts you in the room you specify
                            name = ""
                            for item in split[2:]:
                                name += item + " "
                            card.move(name.strip(), "random")
                        if(split[0] == "move" and len(split)==2):
                            #makes you move in the direction you specify
                            card.move("none", split[1])
                        if(split[0] == "move" and len(split)==3 and not split[1]=="to"):
                            #to move an n number of times
                            for x in range(int(split[2])):
                                card.move("none", split[1])
                        if(split[0] == "add" and len(split)>=3):
                            #adds a room with the direction as the label
                            name = ""
                            for item in split[1:-1]:
                                name += item + " "
                            card.addRoom(name.strip(), split[-1], split[-1], False)
                        if(split[0] == "addl" and len(split)>=4):
                            #adds a room with a unique label
                            name = ""
                            for item in split[1:-2]:
                                name += item + " "
                            card.addRoom(name.strip(), split[-2], split[-1], False)
                        if(split[0] == "addb" and len(split)>=2):
                            #adds a blue room
                            name = ""
                            for item in split[1:]:
                                name += item + " "
                            card.addRoom(name.strip(), "alone", "none", True)
            case 5:
                #exit the program
                break
            case _:
                print("You input an invalid value")
image.save("file.png")
invert_im = ImageOps.invert(image)
bbox = invert_im.getbbox()
cropped = image.crop(bbox)
cropped.save("cropped.png")
```

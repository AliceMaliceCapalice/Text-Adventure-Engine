```python
from PIL import Image, ImageDraw
#add a label to the lines connecting the rooms, mentioning the direction
#I can't seem to use the room move feature.
class Mapping:
    def __init__(self, geometry):
        self.coords = [geometry//2,geometry//2]
        self.rooms = {}
    def addRoom(self, name, direction, label="none"):
        if direction=="start":
            Mapping.drawRoom(self.coords, name)
            self.rooms[name] = self.coords
            return
        self.move("none", direction)
        self.rooms[name] = self.coords
        Mapping.drawRoom(self.coords, name)
        Mapping.drawLine(self.coords, direction)
        if not label=="none":
            Mapping.addLabel(self.coords, direction, label)
    def move(self, destination, direction):
        if not destination == "none":
            #checks if there is a destination
            self.coords=self.rooms[destination]
            return
        if(direction == "north"):
            self.coords = [self.coords[0], self.coords[1]-150]
        if(direction == "south"):
            self.coords = [self.coords[0], self.coords[1]+150]
        if(direction == "east"):
            self.coords = [self.coords[0]+150, self.coords[1]]
        if(direction == "west"):
            self.coords = [self.coords[0]-150, self.coords[1]]
    @staticmethod
    def addLabel(coords, direction, text):
        #This labels the line
        if(direction == "north"):
            draw.text((coords[0], coords[1]+70), text, fill="black")
        if(direction == "south"):
            draw.text((coords[0], coords[1]-70), text, fill="black")
        if(direction == "east"):
            draw.text((coords[0]-70, coords[1]), text, fill="black")
        if(direction == "west"):
            draw.text((coords[0]+70, coords[1]), text, fill="black")
    @staticmethod
    def drawRoom(coords, text):
        draw.rectangle(xy=([(coords[0]-40, coords[1]-40),(coords[0]+40,coords[1]+40)]), outline=(0,0,0))
        draw.text((coords[0]-30, coords[1]), text, fill="black")
    @staticmethod
    def drawLine(coords, direction):
        if direction == "north":
            draw.line([(coords[0],coords[1]+40),(coords[0],coords[1]+110)], fill="black", width=1)
        if direction == "south":
            draw.line([(coords[0],coords[1]-40),(coords[0],coords[1]-110)], fill="black", width=1)
        if direction == "east":
            draw.line([(coords[0]-40,coords[1]),(coords[0]-110,coords[1])], fill="black", width=1)
        if direction == "west":
            draw.line([(coords[0]+40,coords[1]),(coords[0]+110,coords[1])], fill="black", width=1)
#change this line to alter the image size
geometry = 2000
image = Image.new("RGB", (geometry,geometry), (255, 255, 255))
draw = ImageDraw.Draw(image)

card = Mapping(geometry)
name = input("What is the starting location called?")
card.addRoom(name, "start")
while True:
    option = int(input("What would you like to do?\n 1. move.\n 2. add a room adjacent to the current room.\n 3. List rooms. \n 4. to quit.\n"))
    match option:
        case 1:
            #Move to direction
            option = int(input("\nDo you want to 1. go somewhere, or 2. move in a direction"))
            match option:
                case 1:
                    #changing the current room
                    destination = input("Where do you want to go?")
                    card.move(destination, "random")
                case 2:
                    #moving in a direction
                    direction = input("What is the direction you want to move?")
                    card.move("none", direction)
                case _:
                    print("You entered an invalid input")
        case 2:
            #add adjacent room
            name = input("What is the next room called?")
            direction = input("What is the direction of this room?")
            label = input("What do you want the label of the line to be? If you don't want it, type none")
            card.addRoom(name, direction, label)
        case 3:
            #list rooms
            for num, room in enumerate(card.rooms.keys()):
                print(f"\n{num}. {room}")
        case 4:
            #exit the program
            break
        case _:
            print("You input an invalid value")

image.save("File.png")
```
